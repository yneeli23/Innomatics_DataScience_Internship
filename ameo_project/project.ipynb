The dataset was released by Aspiring Minds from the Aspiring Mind Employment Outcome 2015 (AMEO). The study is primarily limited  only to students with engineering disciplines. The dataset contains the employment outcomes of engineering graduates as dependent variables (Salary, Job Titles, and Job Locations) along with the standardized scores from three different areas – cognitive skills, technical skills and personality skills. The dataset also contains demographic features. The dataset  contains  around  40 independent variables and 4000 data points. The independent variables are both continuous and categorical in nature. The dataset contains a unique identifier for each candidate. Below mentioned table contains the details for the original dataset.

```python
fig, ax = plt.subplots(nrows=9, ncols=3, figsize=(20,20))

for i, col in enumerate(num_df.columns):
    row = i // 3
    col = i % 3
    sns.histplot(data=df, x=num_df.columns[i], ax=ax[row, col], kde=True)

fig.tight_layout()

plt.show()
```

BIVARIATE PLOTS

```python
sns.scatterplot(data=df, x=df['collegeGPA'], y=df.English, hue=df['Gender'])

plt.xlabel('GPA')
plt.ylabel('English')
plt.title('Scatter plot of GPA and English')

plt.show()
```

The above plot can be intrepeted in such a way that the more the GPA, the better the peoples english score. We can deduce that there are fewer females and males

```python
plt.hexbin(df.extraversion, df['Quant'], gridsize=25)

plt.xlabel("extraversion")
plt.ylabel("Quant")
plt.title("Hexbin Plot of Quant and Extraversion")

plt.show()
```

We can see that as extraversion rate rises, people scores better quant reasonably

```python
sns.catplot(x="Gender", hue="Specialization", kind="count", data=df)
plt.title("Specialization VS Gender")
plt.show()
```

In above plot, it can deduced that there are majorly computer related specializations

```python
# Create a subset of the data for Computer Science Engineering graduates
d_df = df[df['Degree'] == 'B.Tech/B.E.']

# Filter the subset for job roles of interest
roles = ['Programming Analyst', 'Software Engineer', 'Hardware Engineer', 'Associate Engineer']
d_df = d_df[d_df['Designation'].isin(roles)]

# Calculate the average salary for the subset
avg_salary = d_df['Salary'].mean()

# Compare the average salary with the claimed range
if avg_salary >= 2_50_000 and avg_salary <= 3_00_000:
    print("The TOI claim is justified by given data.")
else:
    print("The TOI claim is not justified by given data.")
```
